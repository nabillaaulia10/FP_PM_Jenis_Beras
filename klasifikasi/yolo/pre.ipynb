```python
from ultralytics import YOLO
import os


# load model
model = YOLO(
    "runs/classify/hasil_training/beras_cls/weights/best.pt"
)


folder_test = "../dataset/test"


# ambil semua gambar di dalam folder test + subfolder kelas
for kelas_asli in os.listdir(folder_test):

    folder_kelas = os.path.join(
        folder_test,
        kelas_asli
    )


    if os.path.isdir(folder_kelas):

        for file in os.listdir(folder_kelas):

            if file.lower().endswith((".jpg",".jpeg",".png")):


                path_gambar = os.path.join(
                    folder_kelas,
                    file
                )


                hasil = model.predict(
                    source=path_gambar,
                    imgsz=224,
                    verbose=False
                )


                for h in hasil:

                    prediksi = h.names[h.probs.top1]

                    confidence = h.probs.top1conf.item()


                    print("====================")
                    print("File :", file)
                    print("Label Asli :", kelas_asli)
                    print("Prediksi :", prediksi)
                    print(
                        "Confidence :",
                        round(confidence*100,2),
                        "%"
                    )
```

```python
from ultralytics import YOLO
import os
import csv


# load model terbaik
model = YOLO(
    "runs/classify/hasil_training/beras_cls/weights/best.pt"
)


folder_test = "../dataset/test"


# folder penyimpanan hasil
output_folder = "runs/test_hasil_beras"

os.makedirs(output_folder, exist_ok=True)


# file laporan hasil
csv_file = os.path.join(
    output_folder,
    "hasil_prediksi.csv"
)


data_hasil = []


# baca semua folder jenis beras
for kelas_asli in os.listdir(folder_test):

    folder_kelas = os.path.join(
        folder_test,
        kelas_asli
    )


    if os.path.isdir(folder_kelas):

        for file in os.listdir(folder_kelas):

            if file.lower().endswith(
                (".jpg",".jpeg",".png")
            ):

                path_gambar = os.path.join(
                    folder_kelas,
                    file
                )


                hasil = model.predict(
                    source=path_gambar,
                    imgsz=224,
                    save=True,
                    project=output_folder,
                    name="gambar"
                )


                for h in hasil:

                    prediksi = h.names[h.probs.top1]

                    confidence = h.probs.top1conf.item()


                    data_hasil.append([
                        file,
                        kelas_asli,
                        prediksi,
                        round(confidence*100,2)
                    ])



# simpan csv

with open(
    csv_file,
    "w",
    newline=""
) as f:

    writer = csv.writer(f)

    writer.writerow([
        "Nama File",
        "Label Asli",
        "Prediksi",
        "Confidence"
    ])

    writer.writerows(data_hasil)



print("Selesai")
print("Hasil tersimpan di:")
print(output_folder)
```
