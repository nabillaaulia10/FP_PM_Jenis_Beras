```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
dataset_path = "dataset_awal/images"

classes = sorted(os.listdir(dataset_path))
```

```python
# JUMLAH DATA PER KELAS

jumlah = []

for cls in classes:
    folder = os.path.join(dataset_path, cls)

    count = len([
        f for f in os.listdir(folder)
        if f.lower().endswith(('.jpg','.jpeg','.png'))
    ])

    jumlah.append(count)

print("\nJumlah Data Tiap Kelas")
print("-"*30)

for c,j in zip(classes,jumlah):
    print(c,":",j)
```

```python
# GRAFIK DISTRIBUSI

plt.figure(figsize=(8,5))

plt.bar(classes,jumlah)

plt.title("Distribusi Dataset Beras")
plt.xlabel("Jenis Beras")
plt.ylabel("Jumlah Gambar")

plt.xticks(rotation=45)

plt.show()
```

```python
# VISUALISASI SAMPEL

fig, ax = plt.subplots(1,len(classes),
                       figsize=(15,5))

for i,cls in enumerate(classes):

    folder = os.path.join(dataset_path,cls)

    img_name = os.listdir(folder)[0]

    img_path = os.path.join(folder,img_name)

    img = cv2.imread(img_path)
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

    ax[i].imshow(img)
    ax[i].set_title(cls)
    ax[i].axis("off")

plt.tight_layout()
plt.show()
```
