```python
from ultralytics import YOLO

model = YOLO(
    "runs/classify/hasil_training/beras_cls/weights/best.pt"
)

hasil = model.predict(
    source="gambar_uji.jpg",
    imgsz=224
)

for h in hasil:

    kelas = h.names[h.probs.top1]
    skor = h.probs.top1conf.item()

    print("Prediksi :", kelas)
    print("Confidence :", round(skor*100,2), "%")
```
