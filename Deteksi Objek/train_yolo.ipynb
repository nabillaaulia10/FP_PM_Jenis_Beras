```python
from ultralytics import YOLO

# ==========================
# LOAD MODEL
# ==========================
model = YOLO("runs/detect/train/weights/best.pt")

# ==========================
# VALIDASI MODEL
# ==========================
metrics = model.val(data="data.yaml")

# ==========================
# HASIL UMUM
# ==========================
print("\n===== HASIL EVALUASI MODEL BERAS =====")

print("mAP@0.5       :", metrics.box.map50)
print("mAP@0.5:0.95   :", metrics.box.map)
print("Precision (avg):", metrics.box.mp)
print("Recall (avg)   :", metrics.box.mr)

# ==========================
# CLASS NAME
# ==========================
names = model.names  
# contoh: {0: 'IR42', 1: 'IR64', 2: 'ketan', 3: 'pandan'}

print("\n===== HASIL PER KELAS BERAS =====")

# ==========================
# PER CLASS METRICS
# ==========================
for i in range(len(names)):
    print(f"\nKelas: {names[i]}")
    print("  Precision :", metrics.box.p[i])
    print("  Recall    :", metrics.box.r[i])
    print("  mAP@0.5   :", metrics.box.ap50[i])
    print("  mAP@0.5:95:", metrics.box.ap[i])
```
